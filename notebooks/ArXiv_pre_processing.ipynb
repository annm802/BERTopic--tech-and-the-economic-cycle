### **Data pre-processing**

```python
#Importing data
import pandas as pd
df= pd.read_csv('/content/drive/MyDrive/cs_papers_api.csv')
df.head(3)
```

```python
# Changing the year column from string to datetime format using functions in pre_processing.py
df= string_to_datetime(df, 'year')

# Keeping instances that come after 2004
df= date_range_filter(df, "year", "2005-01-01" )

#sorting by year and reseting index
cs= df.sort_values(by='year')
cs = cs.reset_index(drop= True)

# Keeping the necessary columns
cs= cs[['title','abstract', 'year']]
cs.head(3)
```

```python
# Train Test split
train_set = cs.sample(frac = 0.975, random_state= 34)

# Creating dataframe with the rest of the 2.5% values
val_set = cs.drop(train_set.index)
```

```python
train_set = train_set.reset_index(drop= True)
val_set = val_set.reset_index(drop= True)
```
